Start with tokenize the text into words

```
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
sentences
```

Calculate the frequency of words in each sentence

```
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

def _create_frequency_matrix(sentences):
  #matrix to hold the frequency of each word in all sentences
  frequency_matrix = {}

  #stopwords to be removed from the text
  stopWords = set(stopwords.words('english'))

  ps = PorterStemmer()

  for sent in sentences:

    #matrix to count the frequency of each word in one sentence
    freq_table = {}

    #Tokenize the sentence into a list of words
    words = word_tokenize(sent)

    for word in words:

      #transform words into lower case
      word = word.lower()

      #removal of suffices, like “ing”, “ly”, “s”, etc
      word = ps.stem(word)

      #not include the word in the table if it is a stop word
      if word in stopWords:
        continue

      #increase the count if the word is already in the table
      if word in freq_table:
        freq_table[word] +=1

      #add that word to the table if it is not in there
      else:
        freq_table[word] = 1

    #add the frequency of words in each sentence into a text matrix
    frequency_matrix[sent[:15]] = freq_table

  return frequency_matrix

  
```

Calculate the TermFrequency for each word in a sentence

**TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)**

The document is a paragraph, while the term is a word in a paragraph. The more frequent a word is in a sentence, the more weight it has

```
def _create_tf_matrix(frequency_matrix):
  
  #matrix to hold the frequency of each word in all sentences
  tf_matrix = {}

  for sent, freq_table in frequency_matrix.items():
    #matrix to hold the frequency of each word in each sentence

    tf_table = {}

    #Calculate the number of words in each sentence
    count_words = len(freq_table)

    for word, count in freq_table.items():

      tf_table[word] = count/count_words

    tf_matrix[sent] = tf_table

  return tf_matrix
```

Create a table for documents per word

```
#matrix to hold the number of documents that hold a particular word
def _create_documents_per_word(frequency_matrix):
  doc_per_word_table = {}

  for sent, freq_table in frequency_matrix.items():
    for word, count in freq_table.items():

      #add the count of document for a word
      if word in doc_per_word_table:
        doc_per_word_table[word] += 1

      else:
        doc_per_word_table[word] = 1

  return doc_per_word_table
```

Calculate IDF for each word in a paragraph

**IDF(t) = log_e(Total number of documents / Number of documents with term t in it)**

The higher the number of sentences with a particular word, the less weight that word has. In another word, since the common word can be trivial, we put penalty for those words

```
import math

def _create_idf_matrix(frequency_matrix, count_doc_per_words, total_documents):
  idf_matrix = {}

  for sent, freq_table in frequency_matrix.items():
    idf_table = {}

    for word in freq_table.keys():
      idf_table[word] = math.log10(total_documents/float(count_doc_per_words[word]))

    idf_matrix[sent] = idf_table

  return idf_matrix
```

Calculate TF-IDF

**TF-IDF algorithm is made of 2 algorithms multiplied together.**

```
def _create_tf_idf_matrix(tf_matrix, idf_matrix):
  tf_idf_matrix = {}

  for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), idf_matrix.items()):

    tf_idf_table = {}

    for (word1, value1), (word2, value2) in zip(f_table1.items(),
                                                f_table2.items()):
      
      tf_idf_table[word1] = float(value1 * value2)
    
    tf_idf_matrix[sent1] = tf_idf_table

  return tf_idf_matrix
```

Score the sentence with Tf_IDF

```
def _score_sentences(tf_idf_matrix) -> dict:
    """
    score a sentence by its word's TF
    Basic algorithm: adding the TF frequency of every non-stop word in a sentence divided by total no of words in a sentence.
    :rtype: dict
    """

    sentenceValue = {}

    for sent, f_table in tf_idf_matrix.items():
        total_score_per_sentence = 0

        count_words_in_sentence = len(f_table)
        for word, score in f_table.items():
            total_score_per_sentence += score

        #Calculate the average score of a sentence
        sentenceValue[sent] = total_score_per_sentence / count_words_in_sentence

    return sentenceValue
```

Find the average score of all sentences as a threshold

```
def _find_average_score(sentence_scores) -> int:

  sumValues = 0

  for entry in sentence_scores:
    sumValues += sentence_scores[entry]

  average = (sumValues / len(sentence_scores))

  return average
```

Select a sentence for a summarization if the sentence score is more than the average score.

```
def _generate_summary(sentences, sentence_scores, average):

    summary = ''

    for sentence in sentences:
        if sentence[:15] in sentence_scores and sentence_scores[sentence[:15]] >= (average):
            summary += " " + sentence

    return summary
```

Put everything together

```
summaries = []
for n in range(len(description_box)):

  text = description_box[n].text

  sentences = sent_tokenize(text)

  #Calculate the frequency of words in each sentence
  frequency_matrix = _create_frequency_matrix(sentences)

  #Calculate the TermFrequency for each word in a sentence
  tf_matrix = _create_tf_matrix(frequency_matrix)

  #number of sentences in the text
  total_documents = len(sentences)

  #calculate the number of documents per word
  count_doc_per_words = _create_documents_per_word(frequency_matrix)

  #Calculate IDF for each word in a paragraph
  idf_matrix = _create_idf_matrix(frequency_matrix, count_doc_per_words, total_documents)

  #Calculate TF-IDF
  tf_idf_matrix = _create_tf_idf_matrix(tf_matrix, idf_matrix)

  #Score the sentence with Tf_IDF
  sentence_scores = _score_sentences(tf_idf_matrix)

  #Find the average score of all sentences as a threshold
  average = _find_average_score(sentence_scores)

  summaries.append(_generate_summary(sentences, sentence_scores, average))
```
